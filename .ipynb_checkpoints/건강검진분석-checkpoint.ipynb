    - 복부비만 : 허리둘레가 남자 90cm 이상, 여자 85cm 이상,

    - 높은 중성지방혈증 : 중성지방(트리글리세라이드) 150mg/dL 이상
    - 낮은 HDL 콜레스테롤혈증 : HDL 콜레스테롤이 남자 40mg/dL미만, 여자 50mg/dL미만

```python
data = pd.read_csv(r"국민건강보험공단_건강검진정보_20211229.CSV", encoding='cp949')
data.info()
```

```python
df = data.dropna()
df.info()
```

```python
df.drop(['기준년도', '가입자 일련번호', '데이터 공개일자'], axis=1, inplace=True)
df.info()
```

```python
# BMI 구하기
df['BMI'] = round(df['체중(5Kg 단위)']/(df['신장(5Cm단위)']/100**2),2)
df.head()
```

```python
# 비만도 검사
df['비만도'] = np.where(df['BMI']>=25,1,0)
df.head()
```

```python
# 복부비만 : 허리둘레가 남자 90cm 이상 여자 85cm 이상
# 결과 : 0, 1
df['복부비만'] = np.where(['성별코드']==1, np.where(df['허리둘레']>=90,1,0), np.where(df['허리둘레']>=85,1,0))
df.head()
```

```python
# 높은 혈압 : 수축기 혈압 130mmHg 이상 또는 이완기혈압 85mmHg 이상
# 결과 : 0, 1
df['혈압'] = np.where((df['수축기 혈압']>=130) | (df['이완기 혈압']>=85),1,0)
df.head()
```

```python
# 높은 혈당 : 공복 혈당 100mg/dL 이상
# 결과 : 0, 1
df['혈당'] = np.where(df['식전혈당(공복혈당)']>=100, 1, 0)
df.head()
```

```python
# 높은 중성지방혈증 : 중성지방(트리글리세라이드) 150mg/dL 이상
# 결과 : 0, 1
df['중성지방혈증'] = np.where(df['트리글리세라이드']>=150, 1, 0)
df.head()
```

```python
# 낮은 HDL 콜레스테롤혈증 : HDL 콜레스테롤이 남자 40mg/dL미만, 여자 50mg/dL미만
# 결과 : 0, 1 HDL 콜레스테롤
df['콜레스테롤혈증'] = np.where(df['성별코드']==1, np.where(df['HDL 콜레스테롤']<40,1,0), np.where(df['HDL 콜레스테롤']<50,1,0))
df.head()
```

```python
# 대사 증후군 합계
df['대사 증후군계'] = df.iloc[:, -5:].sum(axis=1)
df.head()
```

```python
#4.대사 증후군의 범주 결정 
#정상' : 0, '주의군' : 1, '위험군' : 2
df['증후군'] = np.where(df['대사 증후군계']==0, 0, np.where(df['대사 증후군계']<=2, 1, 2))
df.head()
```

##### 복부비만

```python
feature = df[['성별코드', '연령대 코드(5세단위)', '신장(5Cm단위)', '체중(5Kg 단위)', '허리둘레',
       '시력(좌)', '시력(우)', '청력(좌)', '청력(우)', '수축기 혈압', '이완기 혈압', '식전혈당(공복혈당)',
       '총 콜레스테롤', '트리글리세라이드', 'HDL 콜레스테롤', 'LDL 콜레스테롤', '혈색소', '요단백',
       '혈청크레아티닌', '(혈청지오티)AST', '(혈청지오티)ALT', '감마 지티피', '흡연상태', '음주여부',
       '구강검진 수검여부', '치아우식증유무', '치석']]

target = df['복부비만']
```

```python
sns.countplot(x='성별코드', hue = '복부비만', data = df)
plt.show()
```

```python
sns.countplot(x='치석', hue = '복부비만', data=df)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature, target, shuffle = True, stratify = target, random_state = 156)
```

```python
model = DecisionTreeClassifier() 
model.fit(X_train, y_train)
pred = model.predict(X_test)
accuracy_score(y_test, pred)
```

```python
model_dic = {
    "Nearest Neighbors" : KNeighborsClassifier(3),
    "SVM" : SVC(), 
    "Decision Tree" : DecisionTreeClassifier(),
    "Random Forest" : RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1), 
    "LogisticRegression" : LogisticRegression(), 
    "AdaBoost" : AdaBoostClassifier(),
    "Naive Bayes" : GaussianNB(),
    "QDA" : QuadraticDiscriminantAnalysis()
}

model_p = {}
model_acc = {} 
model_auc = {} 

for key in model_dic.keys() :
  model = model_dic[key]
  print(f'{key} : training') 
  model.fit(X_train, y_train)
  
  print(f'{key} : predict') 
  model_p[key] = model.predict(X_test)
  model_acc[key] = accuracy_score(y_test, model_p[key])  
  model_auc[key] =  roc_auc_score(y_test, model_p[key])

result = pd.DataFrame()
result['모델'] = model_acc.keys() 
result['정확도'] = model_acc.values() 
result['AUC'] = model_auc.values()
result.sort_values('정확도', ascending=False)
```
